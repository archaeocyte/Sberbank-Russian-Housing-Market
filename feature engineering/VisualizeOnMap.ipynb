```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplleaflet
```

```python
train = pd.read_csv( '../input/train.csv' )
test = pd.read_csv('../input/test.csv')
```

```python
NAME_to_subarea = pd.read_csv('../input/NAME_to_subarea.csv')
NAME_to_subarea.head()
```

```python
train = train.merge(NAME_to_subarea,on='sub_area')
```

```python
train['OKRUG'].value_counts()
```

```python
colormap = {'Novomoskovsky': 'r', 'South-Eastern' : 'y', 'Southern' : 'b', 'Eastern' : 'c', 
            'South-Western':'w', 'Northern':'k', 'North-Eastern': [0,0.5,1], 'Western':[0.5, 1, 0], 
            'North-Western': [0.5, 0.5, 0.9], 'Central':[1, 0, 1],
            'Zelenogradsky': [1, 0.1, 0.4], 'Troitsky': [0.2, 0.3, 0.4]}
color = []
for data in train['OKRUG']:
     color.append(colormap[data])
```

```python
geoTrain = pd.read_csv( '../input/train_lat_lon.csv' )
joinedTrain = train.join( geoTrain.set_index('id'), on = 'id')
skippedTrain = joinedTrain[::50]

colormap = {'Novomoskovsky': 'r', 'South-Eastern' : 'y', 'Southern' : 'b', 'Eastern' : 'c', 'South-Western':'w', 'Northern':'k',
           'North-Eastern': [0,0.5,1], 'Western':[0.5, 1, 0], 'North-Western': [0.5, 0.5, 0.9], 'Central':[1, 0, 1],
           'Zelenogradsky': [1, 0.1, 0.4], 'Troitsky': [0.2, 0.3, 0.4]}
color = []
for data in skippedTrain['OKRUG']:
     color.append(colormap[data])
        
plt.rcParams["figure.figsize"] = (10, 10)
ax = skippedTrain.plot.scatter( x = 'lon', y = 'lat', s = skippedTrain.price_doc / 1e5, c = color)
mplleaflet.display( fig = ax.figure, closefig = True )
```
